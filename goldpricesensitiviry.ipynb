```python

# ==========================================
# 1. DATA COLLECTION & PREPARATION
# ==========================================
# Load the dataset (Ensure 'gld_price_data.csv' is in your directory)
gold_data = pd.read_csv('gld_price_data.csv')

# Convert Date to datetime and check for missing values
gold_data['Date'] = pd.to_datetime(gold_data['Date'])
print(f"Missing Values Check:\n{gold_data.isnull().sum()}\n")

# ==========================================
# 2. EDA (EXPLORATORY DATA ANALYSIS)
# ==========================================
# Correlation Heatmap to show Sensitivity
plt.figure(figsize=(10, 7))
sns.heatmap(gold_data.drop('Date', axis=1).corr(), cbar=True, square=True, 
            fmt='.2f', annot=True, annot_kws={'size':10}, cmap='YlOrBr')
plt.title('Correlation Matrix: Gold vs Other Assets')
plt.show()

# Checking the Distribution of Gold Prices
plt.figure(figsize=(8, 5))
sns.histplot(gold_data['GLD'], color='gold', kde=True)
plt.title('Gold Price Distribution')
plt.show()


# ==========================================
# 4. PREDICTIVE MODELING
# ==========================================
# Split Features (X) and Target (Y)
X = gold_data.drop(['Date', 'GLD'], axis=1)
Y = gold_data['GLD']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

# Model Training: Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, Y_train)

# ==========================================
# 5. MODEL EVALUATION & REPORTING
# ==========================================
# Predict on Test Data
test_data_prediction = model.predict(X_test)

# Metrics
error_score = metrics.r2_score(Y_test, test_data_prediction)
mae = metrics.mean_absolute_error(Y_test, test_data_prediction)

print(f"--- Model Results ---")
print(f"R Squared Error : {error_score:.4f}")
print(f"Mean Absolute Error : {mae:.4f}")

# Feature Importance (Sensitivity Analysis Output)
importances = model.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'Asset': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Visualizing Sensitivity Results
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Asset', data=feature_importance_df, palette='viridis')
plt.title('Sensitivity Analysis: Which Asset Influences Gold the Most?')
plt.show()

# Actual vs Predicted Plot
Y_test_list = list(Y_test)
plt.figure(figsize=(10,6))
plt.plot(Y_test_list, color='blue', label = 'Actual Value', alpha=0.6)
plt.plot(test_data_prediction, color='green', label='Predicted Value', alpha=0.6)
plt.title('Actual Gold Price vs Predicted Gold Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
